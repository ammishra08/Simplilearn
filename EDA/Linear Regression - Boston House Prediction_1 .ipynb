```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Using sklearn package for internal boston dataset
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
boston
```

```python
print(boston.DESCR)
```

```python
boston_df = pd.DataFrame(boston.data, columns = boston.feature_names)
boston_df
```

```python
# inserting target values into MEDV Column
boston_df['MEDV'] = boston.target
```

```python
boston_df
```

#### Features & Target

```python
# RAD is dropped due to multicollinearity , 
X_feature = boston_df.drop(['RAD','MEDV'], axis = 1)

# Target
Y = boston_df['MEDV']
```

#### Train & Test Split

```python
from sklearn.model_selection import train_test_split
# Split Data into train & test randomnly, random_state = just like np.seed()
# test_size = .2 => 80% is train data , 20 % test data
x_train, x_test, y_train, y_test = train_test_split(X_feature, Y, test_size = 0.2, random_state = 3)
```

#### Data Preprocessing
    * Scaling Down the features range into same range.

```python
x_train
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range = (0,1))
# fit => trained with dataset , transform => apply changes
x_train_scaler = scaler.fit_transform(x_train)
x_test_scaler = scaler.transform(x_test)
```

```python
x_train_scaler
```

#### Linear Regression

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
```

```python
# Training of algorithm with train data & train label
lin_reg.fit(x_train_scaler, y_train)
```

```python
# Test Score (score must be closer to 1) - R2 (coefficient of determination)
lin_reg.score(x_test_scaler, y_test)
```

#### Regression Metrics

```python
# Predictions
yhat = lin_reg.predict(x_test_scaler)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# compairing score of actual & predicted
r2_score(y_test, yhat)
```

```python
# MAE
mean_absolute_error(y_test, yhat)
```

```python
# MSE
mean_squared_error(y_test, yhat)
```

```python
# RMSE
# This value must be closer to actual y range
np.sqrt(mean_squared_error(y_test, yhat))
```
